```python
import numpy as np
import matplotlib.pyplot as plt

N = 100

x = np.linspace(-1, 2, N)
y = np.exp(x)

plt.figure()
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$\exp(x)$')
plt.title('Exponential function')

plt.figure()
plt.plot(x, -np.exp(-x))
plt.xlabel('$x$')
plt.ylabel('$-\exp(-x)$')
plt.title('Negative exponential\nfunction')
# To avoid matplotlib text output
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


norm.pdf(x = 1.0, loc=0, scale=1)
```

```python
fig, ax = plt.subplots()

scale = 1.5

x = np.arange(-4, 4, 0.001)
ax.set_title('N(0, $1^s$)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.plot(x, norm.pdf(x, scale = scale), label='stdv=%1f' % scale)
ax.set_ylim(0, 0.45)

plt.show()
```

```python
fig2, ax2 = plt.subplots()
x = np.linspace(-10, 10, 100)

means = [0.0, 1.0, 2.0, 5.0]

for mean in means:
    ax2.plot(x, norm.pdf(x, loc=mean, scale = scale), label='mean=%.1f, stdev=%.1f' % (mean, scale))

scale = 1.5

x = np.arange(-4, 4, 0.001)
ax2.set_title('Some Normal Distributions')
ax2.set_xlabel('x')
ax2.set_ylabel('pdf(x)')
ax2.set_ylim(0, 0.45)
ax2.legend(loc='best', frameon=True)


plt.show()
```

```python

scale = 3.0

set_value = 106.0
tolerance = 3.0

min_graph = 98.0
max_graph = 114.0
overhang = 1.5

ymax = 0.2

fig, ax = plt.subplots()
ax.set_title("Requested Value Max Filter = %.1f" % set_value)
ax.set_xlabel('effective value filter')
ax.set_ylabel('Probability')
ax.grid(True)
ax.set_ylim(0,ymax)

x= np.arange(min_graph,max_graph,0.01)
px= np.arange(min_graph,set_value+tolerance,0.01)

ax.plot(x, norm.pdf(x, loc=set_value, scale=scale))
ax.fill_between(px, norm.pdf(px, loc=set_value, scale=scale), alpha=0.2, color='g')
ax.text(set_value-2.5,ymax/3,"p(success) = %.1f" % norm.cdf(set_value+tolerance, loc=set_value, scale=scale), fontsize=14)
plt.axvline(x = set_value, color = 'black', label = 'target value')
plt.axhline(y = ymax/8, 
            color = 'r', 
            linestyle = '-', 
            xmin=(set_value-min_graph+overhang)/(max_graph-min_graph+2*overhang), 
            xmax=(set_value+tolerance-min_graph+overhang)/(max_graph-min_graph+2*overhang),
            marker='>',
            dash_capstyle = 'round',
            label = 'tolerance %.1f' % tolerance)
ax.text(set_value+0.1,ymax/7,"tolerance\n    =%.1f" % tolerance, fontsize=11)
plt.show()
```

```python
number = 10
min_graph = 80
scale = 3.0
max_graph = min_graph + int(number * scale * 2)
space = int((max_graph - min_graph)/number)
value_population = range(min_graph, max_graph, space)
ymax = 0.2
print(value_population)

page_size = int(0.7*number)
max_results = int(0.8*number)


fig, ax = plt.subplots()
ax.set_title("Sorting Effects")
ax.set_xlabel('effective value filter')
ax.set_ylabel('Probability')
ax.grid(False)
ax.set_ylim(0,ymax)

tolerance = 3.5
overhang = 12

for val in value_population:
    x= np.arange(val-2.5*scale, val+2.5*scale, scale/100)
    ax.plot(x, norm.pdf(x, loc=val, scale=scale))
    plt.axvline(x = val, color = 'grey', linestyle = '--', linewidth = 0.5, label = 'target value')
plt.axvspan(min_graph-tolerance, value_population[page_size]+tolerance, alpha=0.2, color='grey')

px_page = np.arange(value_population[-3]+1*scale, value_population[-3]+3*scale, 0.1)
ax.fill_between(px_page, norm.pdf(px_page, loc=value_population[-3], scale=scale), alpha=0.4, color='r')
px_max_results = np.arange(value_population[-2]-3*scale, value_population[-2]-1*scale, 0.1)
ax.fill_between(px_max_results, norm.pdf(px_max_results, loc=value_population[-2], scale=scale), alpha=0.4, color='r')
#ax.fill_between(px, norm.pdf(px, loc=set_value, scale=scale), alpha=0.2, color='r')
ax.text(
    value_population[0], 
    ymax*0.9,
    'PageSize = %d' % page_size, fontsize=11)
plt.axhline(y = ymax*0.85, 
            color = 'pink', 
            linestyle = '-', 
            xmax=(2*overhang+value_population[page_size]-min_graph)/(max_graph-min_graph+2*overhang),
            xmin= (overhang)/(max_graph-min_graph+2*overhang),
            marker='>',
            dash_capstyle = 'round',
            label = 'page_size %d' % page_size)
plt.axvspan(min_graph-tolerance, value_population[max_results]+tolerance, ymax=0.8, alpha=0.2, color='green')
ax.text(
    value_population[0], 
    ymax*0.75,
    'maxResults = %d' % max_results, fontsize=11)
plt.axhline(y = ymax*0.7, 
            color = 'r', 
            linestyle = '-', 
            xmax=(2*overhang+value_population[max_results]-min_graph)/(max_graph-min_graph+2*overhang),
            xmin= (overhang)/(max_graph-min_graph+2*overhang),
            marker='>',
            dash_capstyle = 'round',
            label = 'page_size %d' % page_size)

#px= np.arange(set_value-tolerance, set_value+tolerance, 0.01)


#ax.fill_between(px, norm.pdf(px, loc=set_value, scale=scale), alpha=0.2, color='g')
#ax.text(set_value-2.5,ymax/3,"p(success) = %.1f" % (norm.cdf(set_value+tolerance, loc=set_value, scale=scale) - norm.cdf(set_value-tolerance, loc=set_value, scale=scale)), fontsize=14)
#
#plt.axhline(y = ymax/8, 
#            color = 'r', 
#            linestyle = '-', 
#            xmin=(set_value-tolerance-min_graph+overhang)/(max_graph-min_graph+2*overhang), 
#            xmax=(set_value-min_graph+overhang)/(max_graph-min_graph+2*overhang),
#            marker='<',
#            dash_capstyle = 'round',
#            label = 'tolerance %.1f' % tolerance)


plt.show()
```
